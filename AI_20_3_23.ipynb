```python
from sklearn.linear_model import Perceptron
from sklearn import datasets
from sklearn.model_selection import train_test_split

from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler

iris = datasets.load_iris()
x = iris.data
y = iris.target
print(x[:5])
print(y[:5])
```

```python
X_train ,X_test , Y_train, Y_test = train_test_split(x, y, test_size = 0.3)
sc = StandardScaler()
sc.fit(X_train)
x_train_std = sc.transform(X_train)
x_test_std = sc.transform(X_test)
print(x_train_std[:5], X_train[:5])
model = Perceptron(max_iter=40, eta0=0.1, random_state=0)
model.fit(x_train_std, Y_train)
y_pred = model.predict(x_test_std)
print(y_pred)
print(Y_test)
print('độ chính xác mô hình:', accuracy_score(Y_test, y_pred))
```

```python
import pandas as pd
data = pd.read_csv('crabs.csv')
#thay thế
#6 đặc tính + loài => giới tính

data["speies"].replace({"B": "0", "O": "1"}, inplace=True) #M=0 O=1
data["sex"].replace({"F": "0", "M": "1"}, inplace=True)
data.head(50)
```

```python
yy = data["sex"].values.tolist()
data.drop('sex', axis = 1).head()
xx = []
for  index, rows in data.iterrows(): 
  list =[rows.speies, rows.fontallip, rows.rearwidth, rows.length, rows.width, rows.depth]
  xx.append(list)
```

```python
X_train ,X_test , Y_train, Y_test = train_test_split(xx, yy, test_size = 0.3)
sc = StandardScaler()
sc.fit(X_train)
x_train_std = sc.transform(X_train)
x_test_std = sc.transform(X_test)
model = Perceptron(max_iter=40, eta0=0.1, random_state=0)
model.fit(x_train_std, Y_train)
y_pred = model.predict(x_test_std)
print(y_pred)
print(Y_test)
print('độ chính xác mô hình:', accuracy_score(Y_test, y_pred))
```
