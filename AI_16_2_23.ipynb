```python
'''
url='http://winterolympicsmedals.com/medals.csv'
data=pd.read_csv(url)

năm 
thành phố diễn ra thế vận hội
môn thể thao

tìm 1 nước 
tổng số huy chương 
bao nhiêu huy chương vàng
vân vân 
'''

import pandas as pd
import numpy as np
url='http://winterolympicsmedals.com/medals.csv'
data=pd.read_csv(url)
data.head(5)
#dataframe=pd.DataFrame(data)
print('Thông tin tổng quát')
year_list=data['Year'].unique() #list all years
year_count=data['Year'].nunique() #count number of year
country_NOC=data['NOC'].unique()  #list all NOC 
medal_total=data['Medal'].count() #total medal
gold_count=data['Medal'].value_counts()['Gold'] #sum gold medal
silver_count=data['Medal'].value_counts()['Silver'] #sum silver
bronze_count=data['Medal'].value_counts()['Bronze'] #sum bronze
sport_total=data['Sport'].nunique() #count sports
sport_list=data['Sport'].unique() #list of sports
print(f'National Olympic Committee (NOC):\n{country_NOC}') 
print(f'Só thế vận hội Olympic mùa đông đã diễn ra từ 1924 đến 2006: {year_count}')
print(f'Các năm diễn ra thế vận hội Olympic mùa đông:\n{year_list}')
print(f'Tổng số huy chương đã trao: {medal_total} \nHuy chương vàng: {gold_count} \nHuy chương bạc: {silver_count} \nHuy chương đồng: {bronze_count}')
print(f'Tổng số môn thể thao: {sport_total} bao gồm: {sport_list}')

```

```python

#filter
NOC_input=year_input=medal_input=city_input=sport_input=None
while not NOC_input: #bắt buộc phải có NOC
  NOC_input=input('\nBộ lọc theo National Olympic Committee (NOC):')
if NOC_input: 
  temp=data[data['NOC']==NOC_input]['City'].unique()
  print(f'Tên các thành phố {NOC_input} tham gia Olympic: \n{temp}')
  city_input=input('\nBộ lọc theo thành phố:')
year_input=input('\nBộ lọc theo năm (Year):')
medal_input=input('\nBộ lọc theo huy chương:')
sportl_input=input('\nBộ lọc theo môn thể thao:')
if year_input: year_input=int(year_input) #convert to INT
#filtered_data=data[(data['NOC']==NOC_input) & (data['Year']==year_input) & (data['Medal']==medal_input) & (data['City']==city_input) & (data['Sport']==sport_input) ]
filter = (data['NOC'] == NOC_input) if NOC_input else True
filter &= (data['Year'] == int(year_input)) if year_input else True
filter &= (data['Medal'] == medal_input.capitalize()) if medal_input else True
filter &= (data['City'] == city_input) if city_input else True
filter &= (data['Sport'] == sport_input) if sport_input else True
filtered_data= data[filter]
print(filtered_data)
```

```python
if not medal_input:
  gold=filtered_data['Medal'].value_counts()['Gold']
  silver=filtered_data['Medal'].value_counts()['Silver']
  bronze=filtered_data['Medal'].value_counts()['Bronze']
  print(f'{NOC_input} đạt số huy chương: {gold} vàng, {silver} bạc, {bronze} đồng', end=' ')
else:
  medal=filtered_data['Medal'].value_counts()[medal_input]
  print(f'{NOC_input} đạt số huy chương {medal} {medal_input}', end=' ')
if city_input:
  print(f'tại thành phố {city_input}', end=' ')
elif  not city_input:
  city=filtered_data['City'].unique()
  print(f'\ntại các thành phố sau {city}')
if sport_input:
  print(f'ở môn thể thao {sport_input}', end=' ')
elif  not sport_input:
  sport=filtered_data['Sport'].unique()
  print(f'\ntrong những môn thể thao {sport}') 
if year_input or city_input:
  year=filtered_data['Year'].unique()
  print(f'ở năm {year}')
elif  not year_input:
  year=filtered_data['Year'].unique()
  print(f'\nở những năm {year}') 
```
